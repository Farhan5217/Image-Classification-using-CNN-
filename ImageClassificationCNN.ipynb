```python
import random
import matplotlib.pyplot as plt
```

```python
import numpy as np
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Conv2D, Dense, Flatten
```

```python
X_train=np.loadtxt(r'C:\Users\usman\Downloads\Compressed\ImageClassifi\Image Classification\input.csv',delimiter=',')
Y_train=np.loadtxt(r'C:\Users\usman\Downloads\Compressed\ImageClassifi\Image Classification\labels.csv',delimiter=',')

X_test=np.loadtxt(r'C:\Users\usman\Downloads\Compressed\ImageClassifi\Image Classification\input_test.csv',delimiter=',')
Y_test=np.loadtxt(r'C:\Users\usman\Downloads\Compressed\ImageClassifi\Image Classification\labels_test.csv',delimiter=',')
```

```python
print(f'Shape of X_train:{X_train.shape}')
print(f'Shape of Y_train:{Y_train.shape}')
print(f'Shape of X_test:{X_test.shape}')
print(f'Shape of Y_test:{Y_test.shape}')
```

```python
X_train=X_train.reshape(len(X_train),100,100,3)
Y_train=Y_train.reshape(len(Y_train),1)
X_test=X_test.reshape(len(X_test),100,100,3)
Y_test=Y_test.reshape(len(Y_test),1)
```

```python
print(f'Shape of X_train:{X_train.shape}')
print(f'Shape of Y_train:{Y_train.shape}')
print(f'Shape of X_test:{X_test.shape}')
print(f'Shape of Y_test:{Y_test.shape}')
```

```python
X_train[1]
```

```python
# We will normalize these values between 0 and 1
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train[1]
```

```python
# this will randomly generate a image from the dataset

rs=random.randint(0,len(X_train))
plt.imshow(X_train[rs,:])
plt.show()
```

```python
model=Sequential([
    Conv2D(32,(3,3),activation='relu',input_shape=(100,100,3)),
    MaxPooling2D((2,2)),
    
    Conv2D(32,(3,3),activation='relu'),
    MaxPooling2D((2,2)),
    
    Flatten(),
    Dense(64,activation='relu'),
    Dense(1,activation='sigmoid')
    
])
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_train,Y_train,epochs=15, batch_size=64)
```

```python
model.evaluate(X_test,Y_test)
```

```python
rs2=random.randint(0,len(Y_test))
plt.imshow(X_test[rs2,:])
plt.show()

y_pred=model.predict(X_test[rs2,:].reshape(1,100,100,3))

print(y_pred)
```
